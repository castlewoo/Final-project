```python
import pandas as pd
```

```python
food_place = pd.read_csv('./data/food_place.csv', index_col = 0)
```

```python
food_place
```

```python
cols = ['도로명주소', '사업장명', '업태구분명', 'X좌표', 'Y좌표']
data = []
```

```python
for idx in food_place.index : 
    tuple_t = []
    tmp = food_place.loc[idx]
    tuple_t.append(str(tmp['도로명주소']))
    tuple_t.append(str(tmp['사업장명']))
    tuple_t.append(str(tmp['업태구분명']))
    tuple_t.append(str(tmp['좌표정보(X)']))
    tuple_t.append(str(tmp['좌표정보(Y)']))
    data.append(dict(zip(cols,tuple_t)))
```

```python
data
```

```python
# json에 포함시킬 meta data 구성
res = {
        'meta':{
            'desc':'영업 상태 및 주소',
            'cols':{
                 '도로명주소':'도로명주소' 
                ,'사업장명':'사업장명' 
                ,'업태구분명':'업태구분명' 
                ,'X좌표':'X좌표' 
                ,'Y좌표':'Y좌표' 
                  },
        }, 'data':data
}
res
```
