```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
spend_data = pd.read_csv('./data/spend_graph_data.csv', encoding ='CP949')
# spend_data['KTO업종중분류명'].unique()
# spend_data
spend_data = spend_data[spend_data['KTO업종중분류명'] == '식음료']
spend_data = spend_data[['기준년월', '소비금액']]
spend_data
```

```python
plt.plot(spend_data['기준년월'], spend_data['소비금액'], 'bo--', label='소비금액')
plt.xlabel('월별')
# plt.xlim([0, 12])
plt.ylabel('소비금액')
plt.ylim([20000000, 90000000])
plt.legend()
plt.title('2023년 월별 외국인 소비금액')
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.show()
```

```python
region_data = pd.read_csv('./data/region_graph_data.csv', encoding ='CP949')
region_data = region_data[region_data['지자체'] == '서울특별시']
region_data = region_data[region_data['방문자 구분'] == '외국인방문자(c)']
region_data = region_data[['기준년월', '방문자수']]
region_data
```

```python
plt.plot(region_data['기준년월'], region_data['방문자수'], 'bo--', label='방문자수')
plt.xlabel('월별')
# plt.xlim([0, 12])
plt.ylabel('방문자수')
plt.ylim([100000, 1500000])
plt.legend()
plt.title('2023년 월별 외국인 방문자수')
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.show()
```
